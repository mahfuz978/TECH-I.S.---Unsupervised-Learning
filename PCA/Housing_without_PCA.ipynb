# **Load Dataset**
* We will use the Housing Prices dataset

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
target = df[['SalePrice']]
```

```python
df.dtypes
```

```python
num_features = df.select_dtypes(exclude = ['object'])
num_features.head()
```

```python
features = num_features.drop(columns=['Id', 'MSSubClass', 'SalePrice'])
```

```python
features.shape
```

```python
features.isnull().sum()
```

```python
# here filling the nan values with the median of that column
for col in features.columns:
    features[col] = features[col].fillna(features[col].median()) 
```

```python
# check to see if it worked
features.isnull().sum()
```

```python
features.describe()
```

```python
import seaborn as sns
sns.heatmap(features.corr());
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms.fit(features)
```

```python
# get scaled data
df_scaled = pd.DataFrame(mms.transform(features))

df_scaled.head()
```

```python
# assign column labels
df_scaled.columns = features.columns
```

```python
df_scaled.head()
```

```python
df_scaled.describe()
```

```python
features = df_scaled
features.head()
```

```python
features.shape
```

```python
target.shape
```

```python
train = pd.concat([features, target],axis = 1)
train.head()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(train, test_size=0.25, random_state=1)
```

```python
train_x = train.drop(['SalePrice'], axis = 1)
train_y = train['SalePrice']

test_x = test.drop(['SalePrice'], axis = 1)
test_y = test[['SalePrice']]
```

```python
print('Dimension of train_x dataframe:',train_x.shape)
print('Dimension of train_y dataframe:',train_y.shape)
print('Dimension of test_x dataframe:',test_x.shape)
print('Dimension of test_y dataframe:',test_y.shape)
```

```python
from sklearn.linear_model import LinearRegression

ln_reg = LinearRegression()
ln_reg.fit(train_x, train_y)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score

df_pred = ln_reg.predict(test_x)
print('Mean Absolute Error of linear regression:',mean_absolute_error(df_pred, test_y))
print('Mean Square Error of linear regression:', mean_squared_error(df_pred, test_y))
print('R_Squared Score of linear regression:', r2_score(df_pred, test_y))
```

```python
predicted = pd.DataFrame(df_pred, columns = ['predicted'])
```

```python
actual = test_y.reset_index(drop='index')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.relplot(
    data=pd.concat([predicted, actual], axis = 1),
);
```
